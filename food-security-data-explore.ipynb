```python
cols_to_keep = [
    'adm0_name', 'adm1_name', 'adm2_name', 'chtype',
    'reference_label', 'reference_year',
    'population', 'phase_class', 'phase1', 'phase2', 'phase3',
    'phase4', 'phase5', 'phase35', 'usethisperiod', 'foodconsumption_phase',
    'livelihoods_phase','nutrition_phase', 'mortality_phase'
]
df = df[cols_to_keep]
```

```python
# Filter rows where column 'X' equals 'MRT'
filtered_df = df[df['adm1_name'] == 'Assaba']

filtered_df = filtered_df[filtered_df['usethisperiod']=='Y']

#filtered_df = filtered_df[filtered_df['chtype']=='projected']
```

```python
filtered_df = filtered_df.drop('usethisperiod', axis=1)
filtered_df = filtered_df.drop('phase_class', axis=1)
filtered_df = filtered_df.drop('phase5', axis=1)
filtered_df = filtered_df.drop('mortality_phase', axis=1)
```

```python
print(filtered_df.info())
```

```python
filtered_df['adm2_name'] = filtered_df['adm2_name'].fillna('unknown')
```

```python
import pandas as pd
import numpy as np

# Define a function to compute the mean excluding zeros and NaNs
def mean_excluding_zeros(series):
    filtered = series[(series != 0) & (~series.isna())]
    return filtered.mean()

# Define a function to compute the mode (most frequent value)
def mode_value(series):
    # Drop NaN values first
    series = series.dropna()
    if series.empty:
        return np.nan
    # Return the first mode (in case of multiple modes)
    mode_series = series.mode()
    return mode_series.iloc[0] if not mode_series.empty else np.nan

# Define the aggregation dictionary
agg_dict = {
    "adm0_name": "first",
    "adm1_name": "first",
    "chtype": "first",
    "reference_label": "first",
    "population": mean_excluding_zeros,
    "phase1": mean_excluding_zeros,
    "phase2": mean_excluding_zeros,
    "phase3": mean_excluding_zeros,
    "phase4": mean_excluding_zeros,
    "phase35": mean_excluding_zeros,
    "foodconsumption_phase": mode_value,
    "livelihoods_phase": mode_value,
    "nutrition_phase": mode_value
}

# Group by the year and aggregate accordingly
# Group by both reference_year and adm1_name
grouped_df = filtered_df.groupby(["reference_year", "adm2_name"]).agg(agg_dict).reset_index()


print(grouped_df)
```

```python
grouped_df=grouped_df.drop("reference_label",axis=1)
```
